```python
# Modules Import 
import pandas as pd
import re
from datetime import datetime
```

Steps for preparing Future data

```python
# Importing Data
data = pd.read_csv('GFDLNFO_07042017.csv')
data
```

```python
#  Filter BANKNIFTY-I.NFO from Ticker column

data1 = data.query("Ticker == 'BANKNIFTY-I'")
data1

```

```python
#  Convert Date and Time into a new datetime column.
data1.insert(1,'data_time', data1['Date'] + ' ' + data1['Time'])
print(data1)
```

```python
# Remove columns: Ticker, Date, Time, Volume, Open Interest.

data1 =data1.drop(['Date', 'Time', 'Ticker', 'Volume', 'Open Interest'], axis=1)
```

```python
# Sort data by datetime.

data1.sort_values(by=['data_time'])
```

Steps for preparing Option data

```python
# Filter BANKNIFTYxxxxxxxxxxCE.NFO & BANKNIFTYxxxxxxxxxxPE.NFO from ticker column.
pattern = 'BANKNIFTY[0-9]+[A-Z]+[0-9]+[A-Z]+'
data2 = data[data['Ticker'].str.contains(pattern, regex=True)]
data2
```

```python
# Extract scrip (35000PE) from ticker and create separate column named scrip
scrip = '[0-9]{5}[A-Z]{2}'
matches = [re.search(scrip, ticker).group() for ticker in data2['Ticker'] if re.search(scrip, ticker)]
data2.insert(1,'scrip', matches )
```

```python
# Extract expiry date (03FEB22) form ticker and create separate column named expiry.

Expire_date = '[0-9]{2}[A-Z]{3}[0-9]{2}'
output_string =[]
matches = [re.search(Expire_date, ticker).group() for ticker in data2['Ticker'] if re.search(scrip, ticker)]
# matches
for i in matches:
    date_object = datetime.strptime(i, "%d%b%y")
    output_string.append(date_object.strftime("%d/%m/%y"))

data2.insert(8,'Expire', output_string)
```

```python
# Convert Date and Time into a new datetime column.
data2.insert(1,'data_time', data2['Date'] + ' ' + data2['Time'])
```

```python
#  Calculate dte = expiry - Date
output_string = []

for i in range(len(data2['Date'])):
    Expire = datetime.strptime(data2['Expire'].iloc[i], "%d/%m/%y")
    Date = datetime.strptime(data2['Date'].iloc[i], "%d/%m/%Y")
    difference = Expire - Date
    output_string.append(difference.days)


data2.insert(10,'dte', output_string)
```

```python
# filter data by minimum dte value
minimum_dte = min(data2['dte'])
data2 = data2[data2['dte'] == minimum_dte]
```

```python
# Remove columns: Ticker, Date, Time, Volume, Open Interest.

data2 =data2.drop(['Date', 'Time', 'Ticker', 'Volume', 'Open Interest'], axis=1)
```

```python
# Sort data by datetime.
data2.sort_values(by=['data_time'])
```
